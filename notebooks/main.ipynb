```python
"""
Download and stores FRP products
"""

from wekeo.download import get_FRP_products

north_america_area = dict(
    west    = -140.51232138616658,
    south   =   33.646156469954164,
    east    =  -48.97389106866001,
    north   =   66.78613522811129
)

# day = "2025-07-24"
day = "2022-07-15"

files = get_FRP_products(
    start_date = day,
    end_date = day,
    area = north_america_area,
)
```

```python
from wekeo.reader import read_FRP_product

first = files[0]

# Read specific variables
variables = [
    'latitude', 
    'longitude',
    'FRP_SWIR',
    'FRP_uncertainty_SWIR',
    'FRP_MWIR', 
    'FRP_uncertainty_MWIR',
    'confidence_SWIR_SAA',  # SAA flag for quality control
    'solar_zenith',
    # 'S8_Fire_pixel_BT',  # This is from a different dimension
]
    
ds = read_FRP_product(first, variables=variables)
```

```python
from wekeo.plot_FRP import plot_FRP_MWIR, plot_FRP_SWIR

# Call the function with the files from cell 1
plot_FRP_SWIR(
    inputs=files, 
    area=north_america_area, 
    area_name="North America", 
    start_date=day, 
    end_date=day
)

plot_FRP_MWIR(
    inputs=files, 
    area=north_america_area, 
    area_name="North America", 
    start_date=day, 
    end_date=day
)
```
